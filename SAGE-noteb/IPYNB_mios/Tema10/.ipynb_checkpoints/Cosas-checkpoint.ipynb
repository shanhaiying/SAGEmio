```python
#La ruina del jugador: Un jugador acude a un casino con 100 euros y acepta el siguiente juego: se lanza una moneda
#y si sale cara gana un euro del casino y si sale cruz paga un euro al casino. Suponemos que el casino dispone de una
#cantidad ilimitada de dinero.

def ruina(n):
    eur = 100;
    for x in xsrange(n):
        if(randint(0,1) == 0):
            eur = eur - 1;
        else:
            eur = eur + 1;
        if(eur <= 50 or eur >= 200):
            return eur;
    return eur;
def probruina(n):
    cont = 0;
    for x in xsrange(n):
        if(ruina(n) == 200):
            cont = cont +1;
    return cont/n;
probruina(10000)
```

```python
def jugada_mh(eleccion):
    me_cambio = 1
    puertas = [0] * 3
    puertas[randint(0,2)] = 1 #(2)
    quedan = puertas #(3)
    del quedan[eleccion] #(4)
    la_otra = (1 in quedan) #(5)
    if me_cambio == 1:
        return la_otra
    else:
        return puertas[eleccion]
```

```python
sum([jugada_mh(1) for int in xrange(100000)])
```

```python
#Ejercicio 5.1
def prob1(n):
    cont = 0
    aux = 0
    for int in xsrange(n):
        for i in xsrange(1000):
            if(randint(0,1)==0):
                cont += 1
            else:
                cont = 0
            if(cont == 10):
                break
        if(cont==10):
            aux += (1/n).n()
    return aux
```

```python
#Ejercicio 5.2
def sombreros(n, nreps):
    p = 0;
    for x in xsrange(nreps):
        L = []
        flag = 0;
        #Llegan y dejan el sombrero
        for i in xsrange(n):
            L.append(i)
        #Salen y cogen uno aleatorio
        for i in xsrange(n):
            tam = len(L)-1
            aux = randint(0, tam)
            if(L[aux] == i):
                flag = 1
                break
            L.pop(aux)
        if(flag==0):
            p += (1/nreps).n()
    return p
                
sombreros(200,10000)
```

```python
#Ejercicio 6.1
def fisica(lon, N):
    L = [5]*lon
    for x in xsrange(N):
        a = randint(0,lon-1)
        b = randint(0,lon-1)
        while (a == b):
            b = randint(0,lon-1)
        if(L[a]>0):
            L[a] = L[a] - 1
            L[b] = L[b] + 1
    return L
```

```python
for i in xsrange(2,7):
    L = fisica(1000, 10^i)
    T = finance.TimeSeries(L)
    show(T.plot_histogram())
```

```python
# Ejercicio 8

def paseos(N):
    pos = 0;
    for x in xsrange(N):
        if(randint(0,1)==0):
            pos -= 1
        else:
            pos += 1
    return pos
```

```python
# 8.1
def promedio(N, nit):
    media = 0
    for x in xsrange(nit):
        pos = paseos(N)
        media += (abs(pos)/nit).n()
    return media

promedio(10000, 1000)
```

```python
# Ejercicio 9

def paseos2d(N):
    pos = [0]*3
    for x in xsrange(N):
        r = randint(0,3)
        if r == 0:
            pos[0] += 1
        elif r == 1:
            pos[0] -= 1
        elif r == 2:
            pos[1] += 1
        else:
            pos[1] -= 1
        if (pos[0]==0 and pos[1]==0):
            pos[2] += (1/N).n()
    return (pos, vuelve)
```
