```python
import string
def encriptar_cesar(C,k):
    L = list(C)
    L1 = map(ord,L)
    L2 = [(item+k)%256 for item in L1]
    C1 = join(map(chr,L2),sep = "")
    return C1
```

```python
encriptar_cesar(texto,5)
```

```python
def encriptar_cesar2(C,k):
    L1 = map(ord,[x.capitalize() for x in list(C)])
    L2 = [(item+k)%256 for item in L1 if (item != 32 and item != 44 and item != 39 and item != 46 and item != 59)]
    C1 = join(map(chr,L2),sep = "")
    return C1
```

```python
C2 = encriptar_cesar2(texto,5);C2
```

```python
print encriptar_cesar2(encriptar_cesar2(texto,5),251)
```

```python
def analisis_frec(T):
    frecuencias = {}
    N = len(T)
    for letra in T:
        if letra in frecuencias:
            frecuencias[letra] += (1/N).n()
        else:
            frecuencias[letra]=(1/N).n()
    return frecuencias
```

```python
def invertir(dict):
    dict_inv = {}
    for key in dict:
        dict_inv[dict[key]] = ord(key)
    return dict_inv
```

```python
dict2 = invertir(analisis_frec(C2));dict2
```
