### ¿Cuántos ceros hay al final de $100!$?

```python
def contar_ceros(n):
    N = factorial(n)
    k=0
    while N%10**k == 0:
        k += 1
    return k
```

```python
time contar_ceros(100)
```

```python
time contar_ceros(365)
```

```python
def contar(n,m):
    N = factorial(n)
    k=0
    while N%m**k == 0:
        k += 1
    return k
```

```python
contar(100,2)
```

```python
contar(100,5)
```

```python
contar(365,2)
```

```python
contar(365,5)
```

## ¿Cuál es el $n$ menor tal que el número de mi DNI está contenido en factorial de $n$?

Supongamos que mi DNI es 12345678.

```python
'1234' in '012345'
```

```python
'124' in '012345'
```

```python
def dni(N):
    k = 1
    C = str(N)
    C1 = str(factorial(k))
    while not C in C1:
        k += 1
        C1 = str(factorial(k))
    return k
```

```python
time dni(12345678)
```

```python
'12345678' in str(factorial(9135))
```

```python
str(factorial(9135)).index('12345678')
```

```python
str(factorial(9135))[11407:11415]
```

## ¿Es cierto que la suma $1! + 2! + 3! + · · · + n!$ no puede ser un cuadrado si $n \ge 4$?

```python
def suma(n):
    suma = 0
    for i in xsrange(1,n+1):
        suma += factorial(i)
    return suma
```

```python
[n for n in srange(4,100) if is_square(suma(n))]
```

```python
[n for n in srange(4,1000) if is_square(suma(n))]
```

```python
[list(factor(suma(n))) for n in srange(4,50)]
```

```python
[n for n in srange(4,50) if not (11,1) in list(factor(suma(n)))]
```

```python
[n for n in srange(4,1000) if not suma(n)%11==0]
```
